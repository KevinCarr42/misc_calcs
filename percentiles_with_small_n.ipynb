```python
import pandas as pd
import numpy as np

percentiles=[0.01,0.1,0.5,0.9,0.99]
```

```python
pd.DataFrame([1,2,3,4,99]).describe(percentiles=percentiles).T
```

```python
pd.DataFrame([1,2,3,4,99]*10).describe(percentiles=percentiles).T
```

```python
pd.DataFrame([49,50,100]).describe(percentiles=percentiles).T
```

```python
def wb_sim(a=1):
    simulation = list()
    for i in range(2, 100):
        p = pd.DataFrame(
            np.random.weibull(a, i)
        ).describe(percentiles=percentiles).T['90%'].iloc[0]
        simulation.append(p)
        
    pd.DataFrame(simulation).plot(figsize=(16,4), legend=False)
```

```python
wb_sim()
```

```python
wb_sim(10)
```

```python
wb_sim(0.1)
```
